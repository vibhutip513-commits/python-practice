```python
1#0. Count Character Frequency
text = "hello"
freq = {}

for char in text:
    freq[char] = freq.get(char, 0) + 1

print(freq)  # Output: {'h': 1, 'e': 1, 'l': 2, 'o': 1}
```

```python
#11. Count Word Frequencypython 
sentence = "apple banana apple"
words = sentence.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

print(freq)  # Output: {'apple': 2, 'banana': 1}
```

```python
#12. Merge Two Dictionaries
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

# Python 3.9+ merge operator
merged_dict = dict1 | dict2
print(merged_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}
```

```python
#13. Find Key with Maximum Value
my_dict = {"a": 10, "b": 50, "c": 30}

max_key = max(my_dict, key=my_dict.get)
print(max_key)  # Output: 'b'
```

```python
#14. Sort a Dictionary by Keys
my_dict = {"c": 3, "a": 1, "b": 2}

sorted_by_keys = dict(sorted(my_dict.items()))
print(sorted_by_keys)  # Output: {'a': 1, 'b': 2, 'c': 3}
```

```python
#15. Sort a Dictionary by Values
my_dict = {"a": 3, "b": 1, "c": 2}

sorted_by_values = dict(sorted(my_dict.items(), key=lambda item: item[1]))
print(sorted_by_values)  # Output: {'b': 1, 'c': 2, 'a': 3}
```

```python
#16. Create a Dictionary from Two Lists
keys = ["name", "age", "city"]
values = ["Alice", 25, "New York"]

combined = dict(zip(keys, values))
print(combined)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}
```

```python
#17. Invert a Dictionary (Swap Keys and Values)
my_dict = {"a": 1, "b": 2, "c": 3}

inverted = {value: key for key, value in my_dict.items()}
print(inverted)  # Output: {1: 'a', 2: 'b', 3: 'c'}
```

```python
#18. Create a Nested Dictionary for Student Records
students = {
    "stud_01": {"name": "John", "grade": "A", "age": 20},
    "stud_02": {"name": "Emma", "grade": "B", "age": 21}
}

print(students["stud_01"]["name"])  # Output: John
```

```python
#19. Find Duplicate Values in a Dictionary
my_dict = {"a": 1, "b": 2, "c": 1, "d": 3, "e": 2}

seen = set()
duplicates = set()

for val in my_dict.values():
    if val in seen:
        duplicates.add(val)
    else:
        seen.add(val)

print(list(duplicates))  # Output: [1, 2]
```
